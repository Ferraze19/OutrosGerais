```python
import pyautogui
import pyperclip
import time
import import_ipynb
import Base_Dados as bd
```

```python
def Cadastrar():
    lf = bd.Criar_lista_filiados()
    lista_filiados = lf.query('SITUAÇÃO == "TÍTULO/NOME IRREGULAR"')
    time.sleep(3)
    for i in lista_filiados.index:
        pyautogui.click(x=1169, y=594)
        time.sleep(0.5)
        pyautogui.click(x=484, y=411)
        pyautogui.doubleClick(x=484, y=411)
        Titulo = str(lista_filiados.at[i, "Título de Eleitor:"]) #Obter iésima entrada de dado de titulo da tabela
        pyautogui.write(Titulo) #Inserir dado obtido linha anterior
        Nome = lista_filiados.at[i, "Nome:"]#Obter iésima entrada de dado de nome da tabela
        pyperclip.copy(str(Nome))
        pyautogui.press('Tab')
        pyautogui.hotkey("Ctrl", "V") #Inserir dado obtido linha anterior
        pyautogui.press('Tab')
        time.sleep(0.5)
        pyautogui.click(x=547, y=315)
        time.sleep(0.5)
        pyautogui.click()
        pyautogui.click(x=892, y=685)
        time.sleep(1)
        pyautogui.click(x=690, y=528)
        time.sleep(1)
        print(Nome)
```

```python
Cadastrar ()
```
